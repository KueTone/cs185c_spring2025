## Making a Plot

With the data in hand, let's go ahead and make a quick plot. First, we will read in the data:

````python
n_rows = 204
n_cols = 300
variables = ['THETA','SALT']
data = np.fromfile(os.path.join(run_dir,'diags','TS_surf_daily_mean','TS_surf_daily_mean.0000002880.data'),'>f4')
data = np.reshape(data,(len(variables),n_rows,n_cols))
````

Next, let's make a plot:

````python
plt.figure()
C = plt.pcolormesh(data[0,:,:],cmap='turbo', vmin=6, vmax=25)
plt.colorbar(C)
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Specify the path to the .bin file
file_path = '/Users/johnathonlu/Documents/SJSU/cs185c_spring2025/Project/input/THETA_IC.bin'

# Open the binary file in read-binary mode
with open(file_path, 'rb') as f:
    # Read the binary data into a NumPy array
    data = np.fromfile(f, dtype=np.float32)

# Create a plot
plt.figure(figsize=(10, 6))
plt.plot(data)

# Add labels and title
plt.xlabel('Sample Index')
plt.ylabel('Data Value')
plt.title('Plot of Binary Data from .bin File')

# Show the plot
plt.show()

# Optionally, save the plot to a file
# plt.savefig('data_plot.png')
````
